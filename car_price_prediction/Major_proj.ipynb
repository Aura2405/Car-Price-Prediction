```python
x=df.drop("Selling_Price",axis=1)
y=df["Selling_Price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=23)
```

```python
x_train.shape,x_test.shape
```

```python
def model_eval(model,x,y):
    y_pred = model.predict(x)
    
    mse = mean_squared_error(y,y_pred)
    print(f"Mean Squared Error = {mse}")
    
    mae = mean_absolute_error(y,y_pred)
    print(f"Mean Abosolute Error = {mae}")
   
    r2 = r2_score(y,y_pred)
    print(f"r2 Score = {r2}")
    
    return "Success"
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(64,input_dim=x_train.shape[1],activation='linear')),
model.add(Dense(32,input_dim=x_train.shape[1],activation='linear')),
model.add(Dense(32,input_dim=x_train.shape[1],activation='linear')),
model.add(Dense(32,input_dim=x_train.shape[1],activation='linear')),
model.add(Dropout(0.5))
model.add(Dense(1,activation='linear'))

model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mse'])
```

```python
model.fit(x_train,y,epochs=150,batch_size=32,validation_split=0.2)
```

```python
y_pred=model.predict(x_test)
```
